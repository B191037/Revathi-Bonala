```python
import matplotlib.pyplot as plt
```

```python
flowers=[42,20,62,17]
glass=[1,6,3,10]
sticks=[10,20,20,30]
```

```python
#plt.plot([10,20,30,40],[3,4,5,7])
plt.figure(figsize=(10,10))
fig=plt.figure()
plt.plot(flowers,glass,marker="o",color='k',linestyle="--",linewidth="5",markeredgecolor="k",markerfacecolor="y")
plt.xlabel("Flowers")
plt.ylabel("Glass")
plt.title("Line Plot")
plt.grid(True)

fig.savefig("a.png")
```

```python
plt.bar(glass,sticks)
plt.xlabel=("glass")
plt.ylabel=("sticks")
```
